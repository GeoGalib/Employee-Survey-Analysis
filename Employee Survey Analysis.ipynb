**EmpID**	Unique identifier for each employee.

**Gender**	Gender of the employee (e.g., Male, Female, Other).

**Age**	Age of the employee.

**MaritalStatus**	Marital status of the employee (e.g., Single, Married, Divorced, Widowed).

**JobLevel**	Job level of the employee (e.g., Intern/Fresher, Junior, Mid, Senior, Lead).

**Experience**	Number of years of work experience the employee has.

**Dept**	Department where the employee works (e.g., IT, HR, Finance, Marketing, Sales, Legal, Operations, Customer Service).

**EmpType**	Type of employment (e.g., Full-Time, Part-Time, Contract).

**WLB**	Work-life balance rating (scale from 1 to 5).

**WorkEnv**	Work environment rating (scale from 1 to 5).

**PhysicalActivityHours**	Number of hours of physical activity per week.

**Workload**	Workload rating (scale from 1 to 5).

**Stress**	Stress level rating (scale from 1 to 5).

**SleepHours**	Number of hours of sleep per night.

**CommuteMode**	Mode of commute (e.g., Car, Public Transport, Bike, Walk, Motorbike).

**CommuteDistance**	Distance traveled during the commute (in kilometers).

**NumCompanies**	Number of different companies the employee has worked for.

**TeamSize**	Size of the team the employee is part of.

**NumReports**	Number of people reported to by the employee (only applicable for Senior and Lead levels).

**EduLevel**	Highest level of education achieved by the employee (e.g., High School, Bachelor, Master, PhD).

**haveOT**	Indicator if the employee has overtime (True/False).

**TrainingHoursPerYear**	Number of hours of training received per year.

**JobSatisfaction**	Rating of job satisfaction (scale from 1 to 5).

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Dept', y='JobSatisfaction', data=data, order=data.groupby('Dept')['JobSatisfaction'].mean().sort_values().index)
plt.title('Average Job Satisfaction by Department')
plt.xlabel('Department')
plt.ylabel('Average Job Satisfaction Score')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(x='Gender', y='JobSatisfaction', data=data)
plt.title('Average Job Satisfaction by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Job Satisfaction Score')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(x='haveOT', y='JobSatisfaction', data=data)
plt.title('Average Job Satisfaction by Overtime Factor')
plt.xlabel('OT True Vs False')
plt.ylabel('Average Job Satisfaction Score')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(x='MaritalStatus', y='JobSatisfaction', data=data)
plt.title('Average Job Satisfaction by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Job Satisfaction Score')
plt.show()
```

```python
# Create age bins
age_bins = np.arange(20, 70, 10)  # adjust the bin size as needed

# Assign each age to a bin
data['AgeBin'] = pd.cut(data['Age'], bins=age_bins, labels=[f'{b}-{b+9}' for b in age_bins[:-1]])

# Plot the data
plt.figure(figsize=(6, 6))
sns.barplot(x='AgeBin', y='JobSatisfaction', data=data, order=data.groupby('AgeBin', observed=False)['JobSatisfaction'].mean().sort_values().index)
plt.title('Average Job Satisfaction by Age Bin')
plt.xlabel('Age Bin')
plt.ylabel('Average Job Satisfaction Score')
plt.show()
```

```python
# Create age bins
training_hours_bins = np.arange(10, 80, 10)  # adjust the bin size as needed

# Assign each age to a bin
data['TrainingBin'] = pd.cut(data['TrainingHoursPerYear'], bins=training_hours_bins, labels=[f'{b}-{b+9}' for b in training_hours_bins[:-1]])

# Plot the data
plt.figure(figsize=(6, 6))
sns.barplot(x='TrainingBin', y='JobSatisfaction', data=data)
plt.title('Average Job Satisfaction by Training Hours')
plt.xlabel('Training hours per year')
plt.ylabel('Average Job Satisfaction Score')
plt.show()
```

**Final Observation**

So it seems like the Job Satisfaction level is only affected by some key factors. The most notable factor here is the marital status where Married people are the least satisfied with their job which is a bit concerning as the the majority of the company are filled with married people. 

People doing overtime are also not satisfied with their job which is to be expected as it is a common scenario. 

Male people are relatively not satisfied with their job compared to female or other gender. 

The two widowed people in the company who are senior citizen are very happy with their current position. 

People age between 40-50 are the most happiest in their current position. 

As of department special care needs to be given to IT, Customer Service and Marketing Department as they are not much happy with their job. 

It seems like upto 60 hours of training per year is enough as anything more than that affect the job satisfaction level.
